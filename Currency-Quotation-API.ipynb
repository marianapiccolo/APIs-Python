Use API to get Currency Quotation information in JSON format.

Open API: https://docs.awesomeapi.com.br/api-de-moedas 


Key Label <br>
bid = aquisição <br>
ask = sale <br>
varBid = variation <br>
pctChange = variation percentage <br>
high = maximum <br>
low = minimum <br>

### Current Price of All Currencies

```python
import requests # request site information
import json # Handle the information in JSON format

quote = requests.get('https://economia.awesomeapi.com.br/json/all')
quote_dic = quote.json() # turn into dictionary
print(quote_dic)
```

### Last quote for Dollar, Euro and BitCoin

```python
print('Dolar: {}'.format(quote_dic['USD']['bid']))
print('Euro: {}'.format(quote_dic['EUR']['bid']))
print('Bitcoin: {}'.format(quote_dic['BTC']['bid']))
```

### Last 30 days

```python
quote_dolar30 = requests.get('https://economia.awesomeapi.com.br/json/daily/USD-BRL/30')
quote_dolar30_dic = quote_dolar30.json()
#print(quote_dolar30_dic[0])

list_quote_dolar = [float(item['bid']) for item in quote_dolar30_dic] # scrolling through the dictionary to get the bid
print(list_quote_dolar)
```

### Bitcoin quote from January 2020 to October 2020

```python
quote_btc = requests.get('https://economia.awesomeapi.com.br/json/daily/BTC-BRL/200?start_date=20200101&end_date=20201031') # 200 quotes
quote_btc_dic = quote_btc.json()
list_quote_btc = [float(item['bid']) for item in quote_btc_dic] 
list_quote_btc.reverse() # reverse the order
print(list_quote_btc)
```

### Bitcoin quotes graphic

```python
# Bitcoin price in 200 days - from January 2020 to October 2020
import matplotlib.pyplot as plt

plt.figure(figsize = (15, 5 ))
plt.plot(list_quote_btc)
plt.show()
```
